```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['lines.markersize'] = 30
```

# Paper Problems

### 1a

```python
x_1 = np.array([1,1,1,1,-1,-1,0,0,-1,-1,3,1]).reshape(4,3)
x_1, y_1 = np.split(x_1, [2], axis=1)
```

```python
pos = np.where(y_1>0)[0]
plt.scatter(x_1[pos,0], x_1[pos,1],marker='+')
neg = np.where(y_1<0)[0]
plt.scatter(x_1[neg,0], x_1[neg,1],marker='_')
# for values of x_1 what is satisfactory x_2? 
x1 = np.linspace(-1,1,100)
x2 = (-2/3)*x1 + (4/3)
plt.plot(x1,x2)
plt.savefig('paper_1a')
```

```python
# 2x1 + 3x2 − 4 = 0
w = np.array([2,3])
w_norm = np.sqrt((w**2).sum())
b = -4
margin = (((x_1 @ w) + b) / w_norm) * y_1[:,0]
```

```python
[print(f'{margin[i]=} for pt {x_1[i]}') for i in range(x_1.shape[0])]
```
